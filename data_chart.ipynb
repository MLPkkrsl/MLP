```python
import pandas as pd

# Load the newly uploaded file to process the updated data
updated_file_path = file_path = r'C:\Users\there\Desktop\SV\project2\2023 등록장애인 현황.xlsx'
updated_data = pd.ExcelFile(updated_file_path)

# Check the sheet names to understand the file structure
updated_data.sheet_names
```

```python
# Load the data from the updated file's first sheet
updated_df = updated_data.parse('Sheet1')

# Display the first few rows of the updated dataset to understand its structure
updated_df.head()

```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정 (예: NanumGothic 또는 다른 한글 폰트)
rc('font', family='Malgun Gothic')  # Windows에서는 'Malgun Gothic', Mac은 'AppleGothic'

# Clean and process the updated data
updated_df.columns = ['장애유형', '합계']  # Ensure consistent column naming
updated_df['합계'] = pd.to_numeric(updated_df['합계'], errors='coerce')  # Convert the '합계' column to numeric

# Prepare data for the pie chart
labels = updated_df['장애유형']
sizes = updated_df['합계']
colors = ['#ff9999', '#99ff99', '#66b3ff', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c4e17f']

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=100,
    colors=colors,
    textprops={'fontsize': 12},
    explode=[0, 0, 0.1, 0, 0, 0, 0]
)

plt.title('장애 유형별 등록 현황', fontsize=16)
plt.show()

# 기타: (호흡기, 심장, 언어, 뇌병변, 신장, 간, 안면)
```
