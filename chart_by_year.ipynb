```python
import pandas as pd

# Load the newly uploaded file to process the updated data
file_path = file_path = r'C:\Users\there\Desktop\SV\project2\연도별 자료.xlsx'
data = pd.ExcelFile(file_path)

# Check the sheet names to understand the file structure
data.sheet_names
```

```python
# Load the data from the updated file's first sheet
df = data.parse('데이터')

# Display the first few rows of the updated dataset to understand its structure
df.head()

```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# Ensure numeric data for the bar chart
df['연도'] = df['연도'].astype(str)  # Convert '연도' to string for proper labeling
df['청각'] = pd.to_numeric(df['청각'], errors='coerce')  # Ensure '청각' is numeric
rc('font', family='Malgun Gothic')  # Windows에서는 'Malgun Gothic', Mac은 'AppleGothic'

# Plot a bar chart for the 청각 data over the years
plt.figure(figsize=(8, 5))
plt.bar(df['연도'], df['청각'], color='skyblue')
plt.title('연도별 청각장애인 등록 현황', fontsize=12)
plt.xlabel('연도', fontsize=12)
plt.ylabel('청각장애인 수', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
